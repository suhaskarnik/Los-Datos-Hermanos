Dataset: Weekly, similar to Smarket, but contains weekly returns for 21 years (1900-2010)

```r
library(ISLR)
cor(Weekly[,-9])
plot(Weekly[,-9])
N=length(Weekly$Direction)
set.seed(1)
```

### 10(a) Log Reg, Direction as response and Lags+Volume as predictors

```r
glm.fit=glm(Direction~Lag1+Lag2+Lag3+Lag4+Lag5+Volume,family=binomial,data=Weekly)
summary(glm.fit)
```

The only statistically significant (p<0.05) are the Intercept and Lag2. Next step, create a Confusion matrix

But as I'm lazy, creating a few functions to help with generating the confusion figures

```r
TPR=function(predicted,actual,positiveVal="Up"){
    tpr=sum(predicted==actual & predicted==positiveVal)/sum(actual==positiveVal)
    round(tpr*100,2)
}

TNR=function(predicted,actual,positiveVal="Up"){
    tnr=sum(predicted==actual & predicted!=positiveVal)/sum(actual!=positiveVal)
    round(tnr*100,2)
}

FPR=function(predicted,actual,positiveVal="Up"){
    fpr=sum(predicted!=actual & predicted==positiveVal)/sum(actual!=positiveVal)
    round(fpr*100,2)
}

FNR=function(predicted,actual,positiveVal="Up"){
    fnr=sum(predicted!=actual & predicted!=positiveVal)/sum(actual==positiveVal)
    round(fnr*100,2)
}

PPV=function(predicted,actual,positiveVal="Up"){
    ppv=sum(predicted==actual & predicted==positiveVal)/sum(predicted==positiveVal)
    round(ppv*100,2)
}

NPV=function(predicted,actual,positiveVal="Up"){
    npv=sum(predicted==actual & predicted!=positiveVal)/sum(predicted!=positiveVal)
    round(npv*100,2)
}

FDR=function(predicted,actual,positiveVal="Up"){
    fdr=sum(predicted!=actual & predicted==positiveVal)/sum(predicted==positiveVal)
    round(fdr*100,2)
}

FOR=function(predicted,actual,positiveVal="Up"){
    f.o.r=sum(predicted!=actual & predicted!=positiveVal)/sum(predicted!=positiveVal)
    round(f.o.r*100,2)
}

accuracy=function(predicted,actual,positiveVal="Up"){
    acc=sum(predicted==actual)/length(actual)
    round(acc*100,2)
}

confusionFigures=function(predicted,actual,name,positiveVal="Up"){
    data.frame(
        Name=name,
        TPR=TPR(predicted,actual,positiveVal),
        TNR=TNR(predicted,actual,positiveVal),
        FPR=FPR(predicted,actual,positiveVal),
        FNR=FNR(predicted,actual,positiveVal),
        PPV=PPV(predicted,actual,positiveVal),
        NPV=NPV(predicted,actual,positiveVal),
        FDR=FDR(predicted,actual,positiveVal),
        FOR=FOR(predicted,actual,positiveVal),
        accuracy=accuracy(predicted,actual,positiveVal),
        TP=sum(predicted==actual & predicted==positiveVal),
        TN=sum(predicted==actual & predicted!=positiveVal),
        FP=sum(predicted!=actual & predicted==positiveVal),
        FN=sum(predicted!=actual & predicted!=positiveVal),
        population=length(predicted)
    );
}

# Simulate tossing a coin
coinToss=function(N,values=c("Up","Down")){
    l=length(values)
    values[ceiling(runif(N)*l)]
}
```

```r
glm.probs=predict(glm.fit,type="response")
glm.pred=rep("Down",N)
glm.pred[glm.probs>0.5]="Up"

# Results of GLM
confFig=confusionFigures(glm.pred,Weekly$Direction,"GLM with all vars")

# Blindly predict an "Up" every week
confFig=rbind(confFig,
                  confusionFigures(rep("Up",N),Weekly$Direction,"Blindly predict Up"))       

# Toss a coin to determine the direction
confFig=rbind(confFig,
                confusionFigures(coinToss(N),Weekly$Direction,"Coin toss"))

confFig
```

### Mistakes made by Log Reg
* FPR is too damn high. A "Down" given by this algorithm is essentially useless
* When we blindly predict an "Up" every day, we have almost the same PPV, Accuracy and False Detection Rate
* Even a coin-toss has a comparable PPV

Let's separate the training and test sets

```r
train=Weekly$Year<2008
NTrain = sum(train)
NTest = N-NTrain
```

Now let's try with only Lag2

```r
glm.fit=glm(Direction~Lag2,family=binomial,data=Weekly,subset=train)
summary(glm.fit)
```

Now Lag2 isn't statistically significant any more

```r
glm.probs=predict(glm.fit,Weekly[!train,],type="response")
glm.pred=rep("Down",length(Weekly$Direction[!train]))
glm.pred[glm.probs>0.5]="Up"

confFig=rbind(confFig,
              confusionFigures(glm.pred,Weekly$Direction[!train],"GLM with only Lag2"))          # Results of GLM

confFig
```

Not much improvement. FPR has improved, but FNR has worsened. Accuracy is no better

# LDA

```r
library(MASS)
```

```r
lda.fit=lda(Direction~Lag2,data=Weekly,subset=train)

lda.pred=predict(lda.fit,Weekly[!train,],type="response")

lda.class=lda.pred$class
cf=confusionFigures(lda.class,Weekly$Direction[!train],"LDA with default threshold")
confFig=rbind(confFig,cf)


for(threshold in seq(0.1,1,by=0.1)){
    lda.class=ifelse(lda.pred$posterior[,"Down"]>=threshold,"Down","Up") 
    cf=confusionFigures(lda.class,Weekly$Direction[!train],paste("LDA with threshold ",threshold))
    confFig=rbind(confFig,cf)
}

confFig
```

Results with LDA are not more promising either

# QDA

```r
qda.fit=qda(Direction~Lag2,data=Weekly,subset=train)
qda.fit
```

# KNN

```r
library(class)
train.X=as.data.frame(cbind(Weekly$Lag2)[train,])
test.X=as.data.frame(cbind(Weekly$Lag2)[!train,])
train.Direction=Weekly$Direction[train]
```

```r
set.seed(1)


#confusionFigures(knn.pred,Weekly$Direction[!train],"KNN K=1")

for(K in seq(1,20)){
    knn.pred=knn(train.X,test.X,train.Direction,k=K)
    cf=confusionFigures(knn.pred,Weekly$Direction[!train],paste("KNN, K=",K))
    confFig=rbind(confFig,cf)
}

confFig
```

```r
cfTop=head(confFig[order(confFig$accuracy,decreasing = TRUE),],5)
plot(cfTop$Name,cfTop$accuracy)
```
